```python
bothchroma_file = '/storage/naumtsevalex/harmony/data/McGill-Billboard/0003/bothchroma.csv'
bothchroma_dir = os.path.dirname(bothchroma_file)
autochord_lab_file = os.path.join(bothchroma_dir, 'autochord.lab')
```

```python
# https://mirdata.readthedocs.io/en/stable/_modules/mirdata/datasets/billboard.html
import numpy as np
import csv

def load_chroma(bothchroma_path):
        """Non-negative-least-squares (NNLS) chroma vectors from the Chordino Vamp plug-in

        Returns:
            np.ndarray - NNLS chroma vector
        """
        # removed the first column since it contains metadata.
        with open(bothchroma_path, "r") as f:
            return np.array([l for l in csv.reader(f)])[:, 1:].astype(np.float32)

```

```python
chroma_vectors = load_chroma(bothchroma_file)
chroma_vectors.shape
# chroma_vectors
chroma_vectors = chroma_vectors[:, 1:]
chroma_vectors.shape
```

```python
labels_lst = autochord.recognize(None, lab_fn=autochord_lab_file, chroma_vectors=chroma_vectors)
```

```python
mcGill_dir = '/storage/naumtsevalex/harmony/data/McGill-Billboard'
for track in tqdm(os.listdir(mcGill_dir)):
    if not track.isdigit():
        continue
    
    track_dir = os.path.join(mcGill_dir, track)
    bothchroma_file = os.path.join(track_dir, 'bothchroma.csv')
    autochord_lab_file = os.path.join(track_dir, 'autochord.lab')
    if os.path.exists(autochord_lab_file):
        continue
    chroma_vectors = load_chroma(bothchroma_file)
    chroma_vectors = chroma_vectors[:, 1:]

    
    labels_lst = autochord.recognize(None, lab_fn=autochord_lab_file, chroma_vectors=chroma_vectors)
    print(autochord_lab_file)
    
```
